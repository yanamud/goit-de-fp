# Завантажувати файл з ftp-сервера в оригінальному форматі csv

```
import requests

# Функція для завантаження файлів
def download_data(local_file_path):
    url = "https://ftp.goit.study/neoversity/"
    downloading_url = url + local_file_path + ".csv"
    print(f"Downloading from {downloading_url}")
    response = requests.get(downloading_url)

    if response.status_code == 200:
        # Зберігаємо файл у тимчасовій директорії Databricks
        with open(f"/tmp/{local_file_path}.csv", 'wb') as file:
            file.write(response.content)
        print(f"File downloaded successfully: /tmp/{local_file_path}.csv")
    else:
        raise Exception(f"Failed to download the file. Status code: {response.status_code}")

# Використання функції
download_data("athlete_bio")
download_data("athlete_event_results")
```
